```python
import pickle
import numpy as np

from matplotlib import pyplot as plt
```

```python
meta = pickle.load(open('../data/graphs/meta.pkl', 'rb'))
```

```python
def process_res(res, key=0):
    final_result, losses, scores = res
    mean_losses = [np.mean(loss) for loss in losses]
    mean_f1 = [score[key]['F1_mean'] for score in scores]
    mean_fm = [score[key]['FM_mean'] for score in scores]
    return mean_losses, mean_f1, mean_fm, final_result[key]['F1_mean'], final_result[key]['FM_mean']
```

```python
def draw_plot(key, results, names, title, ylabel, savename):
    plt.figure(figsize=(8, 6))
    plt.title(title, fontsize=18)
    for res, name in zip(results, names):
        res = process_res(res)
        plt.plot(range(1, len(res[key]) + 1), res[key], label=name)
    plt.grid()
    plt.legend(fontsize=16)
    plt.xlabel('Number of epoches', fontsize=16)
    plt.ylabel(ylabel, fontsize=16)
    plt.savefig(savename, dpi=300)
    plt.show()
```
